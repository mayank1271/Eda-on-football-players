### with using if else we fill the all nan value

```python
for col in data.columns:
    if data[col].dtypes=="object":
        data[col].fillna(data[col].mode()[0],inplace=True)
    else:
        data[col].fillna(data[col].mean(),inplace=True)
        
```

```python
##20.Who are the top 10 eldest players?
```

```python
age=pd.DataFrame(data["Age"])
name=pd.DataFrame(data["Name"])
old=pd.concat(objs=[name,age],axis=1)
old1=old.groupby('Name')[['Age']].max().sort_values(by="Age",ascending=False)
oldest=old1[0:11]
plt.figure(figsize=(12,5))
sns.barplot(oldest,x="Name",y="Age",estimator="count",errorbar=None)
```

```python
##21Who are the top 10 youngest players?
```

```python
young=old.groupby('Name')[['Age']].min().sort_values(by="Age",ascending=True)
young1=young[0:10]
young1.plot(kind="bar",subplots=True,color="pink")
```
